```python
import pandas as pd
import numpy as np
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
# 샘플 생성
sample_size = 10000
sample1 = np.random.normal(loc  = 0.0, scale = 1.0, size = sample_size)
sample2 = np.random.normal(loc  = 0.0, scale = 1.0, size = sample_size)

# scatter
# plt.scatter(sample1,sample2)
```

```python
import matplotlib.gridspec as gridspec

# subplot으로 구역나누기
plt.figure()
gspec = gridspec.GridSpec(3,3)

top_hist = plt.subplot(gspec[0,:2])
side_hist = plt.subplot(gspec[1:,2:])
main_hist = plt.subplot(gspec[1:,:2])

main_hist.scatter(sample1,sample2, color = 'red')
top_hist.hist(sample1, bins=100, normed = True, color = 'blue')
s = side_hist.hist(sample2, bins=100, orientation = 'horizontal', normed = True, color = 'blue')

```
