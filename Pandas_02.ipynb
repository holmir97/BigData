# Dataframe Indexing and Loading

```python
df = pd.read_csv('olympics.csv')
df.head()
```

```python
# 파라미터값을 주어 원하는 형태로 볼 수 있다.
# 아래코드는 column을 하나 제거
df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)
df.head()
```

```python
df.columns
```

```python
# № Summer', '01 !', '02 !', '03 !' 를 원하는 형태로도 변경 가능
# 필요없는 row나 column을 줄이고 데이터를 가져올 수 있따.
for col in df.columns:
    if col[:2]=='01':
            df.rename(columns = {col:'Gold' + col[4:]},inplace=True)
    if col[:2]=='02':
            df.rename(columns = {col:'Silver' + col[4:]},inplace=True)
    if col[:2]=='03':
            df.rename(columns = {col:'Bronze' + col[4:]},inplace=True)
    if col[:1]=='№':
            df.rename(columns = {col:'#' + col[1:]},inplace=True)
df.head()
```

# Querying a Dataframe

```python
# 쿼링 실행_원하는 정보 출력
#Boolean Mask = 내가 원하는 데이터를 꺼내기 위해서 원하는 데이터가 어디있는지 명시하는 것
```

```python
# 금메달을 딴 나라
df['Gold']>0
```

```python
# where을 이용하여 금메달을 딴 나라 출력
only_gold = df.where(df['Gold']>0)
only_gold.head()
```

```python
# 금메달 개수
only_gold['Gold'].count()
```

```python
only_gold.count()
```

```python
# 원래 총 개수
df.count()
```

```python
only_gold.head()
```

```python
# Not a Number(NAN) 지우기
# dropna = na가 있는 것들을 다 제거한다
only_gold = only_gold.dropna()
only_gold.head()
```

```python
# 위와 같은 값을 가지지만, 이 구문에서는 df안에서 진행하게 된다.
# 위와 다르게 float가 아닌 int로 출력
# 위와 다르게 NAN값만 없어지는게 아니라 그 줄 자체가 제거되는것이다.
# 따라서 위의 식에서는 Afghanistan가 살아있지만 여기서는 그것까지 통째로 사라졌다.
only_gold1 = df[df['Gold']>0]
only_gold1.head()
```

```python
# SUMMER와 WINTER모두 금메달이 있는 나라
both_gold = df[(df['Gold']>0) & (df['Gold.1']>0) ]
both_gold.head()
```

```python
len(both_gold)
```

```python
# 금메달을 하나라도 딴 나라
any_gold = df[(df['Gold']>0) | (df['Gold.1']>0)]
len(any_gold)
```

```python
# Stop and resume
# 겨울에 금메달은 따고, 여름엔 따지 못한 나라
df[(df['Gold.1']>0) & (df['Gold']==0) ]
```
