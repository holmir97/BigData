```python
import pandas as pd
```

```python
# 백신 접종 현황
df_vaccine = pd.read_csv('Vaccine.csv')
df_vaccine.head(20)
```

```python
# 칼럽을 확인하여 원하는 데이터만 추출
print(df_vaccine.columns)
```

```python
# 일별 1차 접종자 추출
df_vaccine_ext = df_vaccine[['등록일','1분기 1차실접종자(서울시)']]
```

```python
# 등록일 DATE형식으로 변환
df_vaccine_ext['등록일'] = df_vaccine_ext['등록일'].astype('str')
df_vaccine_ext['등록일'] = pd.to_datetime(df_vaccine_ext['등록일'])
df_vaccine_ext
```

```python
# 시각화
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
plt.figure()
y = df_vaccine_ext['1분기 1차실접종자(서울시)']

df_vaccine_ext['1분기 1차실접종자(서울시)'].plot().color = 'c'

plt.title("Number of coronavirus vaccinations(in Seoul)", fontsize = 15, color = 'red')
plt.ylabel("Number of people")
plt.xlabel("Date (Feb - April)")
```
